```python
def FindTarget(nums, left, right, target):
  mid = int(left + (right - left))

  if right - left > 0:
    if nums[mid] == target:
      return True
    else:
      if nums[mid] < target:
        return FindTarget(nums, mid + 1, right, target)
      elif nums[mid] > target:
        return FindTarget(nums, left, mid - 1, target)

  elif right - left == 0:
    if nums[mid] == target:
      return True
    else:
      return False
  
  else:
    return False
```
