```python
nums = [2,0,2,1,1,0]
nums = [2,0,1]
nums = [0]
```

```python
zeros = 0
ones = 0
twos = 0

for num in nums:
  if num == 0:
    zeros += 1
  elif num == 1:
    ones += 1
  elif num == 2:
    twos += 1

arr = [zeros, ones, twos]


pt = 0

for i in range(len(nums)):
  while arr[pt] == 0:
    pt += 1

  nums[i] = pt
  arr[pt] -= 1
```

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        zeros = 0
        ones = 0
        twos = 0

        for num in nums:
            if num == 0:
                zeros += 1
            elif num == 1:
                ones += 1
            elif num == 2:
                twos += 1

        arr = [zeros, ones, twos]


        pt = 0

        for i in range(len(nums)):
            while arr[pt] == 0:
                pt += 1

            nums[i] = pt
            arr[pt] -= 1
```
