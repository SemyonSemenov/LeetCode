```python
def FindTarget(nums, left, right, target):
  mid = int(left + (right - left))

  if right - left > 0:
    if nums[mid] == target:
      return mid
    else:
      if nums[mid] < target:
        return FindTarget(nums, mid + 1, right, target)
      elif nums[mid] > target:
        return FindTarget(nums, left, mid - 1, target)

  elif right - left == 0:
    if nums[mid] == target:
      return mid
    else:
      return False
  
  else:
    return False

def FindPivot(nums, left, right):
  mid = int(left + (right - left) / 2)
  if right - left > 0:
    if nums[left] < nums[mid]:
      return FindPivot(nums, mid, right)

    elif nums[left] > nums[mid]:
      return FindPivot(nums, left, mid)

  if right - left == 1:
    if nums[left] > nums[right]:
      return right
    else:
      return 'sorted'


class Solution:
    def search(self, nums: List[int], target: int) -> int:
        if len(nums) == 1:
            if nums[0] == target:
                return 0
            else:
                return -1
        
        pivot = FindPivot(nums, 0, len(nums) - 1)
        
        if pivot == 'sorted':
            ans = FindTarget(nums, 0, len(nums) - 1, target)
            if ans is not False:
                return ans
            else:
                return -1
        else:
            ans1 = FindTarget(nums, 0, pivot - 1, target)
            ans2 = FindTarget(nums, pivot, len(nums) - 1, target)
            
            if ans1 is not False:
                return ans1
            elif ans2 is not False:
                return ans2
            else: return -1
```
