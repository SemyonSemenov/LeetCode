```python
def FindTarget(nums, left, right, target):
  mid = int(left + (right - left))

  if right - left > 0:
    if nums[mid] == target:
      return mid
    else:
      if nums[mid] < target:
        return FindTarget(nums, mid + 1, right, target)
      elif nums[mid] > target:
        return FindTarget(nums, left, mid - 1, target)

  elif right - left == 0:
    if nums[mid] == target:
      return mid
    else:
      return False
  
  else:
    return False
```

```python
def FindLeftPos(nums, left, right, curr_pos, target):
  mid = int(left + (right - left))

  if right - left > 0:
    if nums[mid] == target:
      curr_pos = mid
      return FindLeftPos(nums, left, mid - 1, curr_pos, target)

    elif nums[mid] != target:
      return FindLeftPos(nums, mid + 1, right, curr_pos, target)
    
  elif right - left == 0:
    if nums[mid] == target:
      curr_pos = mid
      return curr_pos
    else:
      return curr_pos

  else:
    return curr_pos
```
