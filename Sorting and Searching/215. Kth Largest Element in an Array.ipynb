```python
nums = [3,2,1,5,6,4]
k = 2

nums = [3,2,3,1,2,4,5,5,6]
k = 4
```

```python
nums = sorted(nums)
```

```python
nums[len(nums) - k]
```

```python
class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        nums = sorted(nums)
        return nums[len(nums) - k]
```
