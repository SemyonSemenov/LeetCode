```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        nums = [[0] * n for i in range(m)]
        
        for i in range(n):
            nums[0][i] = 1

        for j in range(m):
            nums[j][0] = 1
            
        for j in range(1, m):
            for i in range(1, n):
                nums[j][i] = nums[j][i - 1] + nums[j - 1][i]
                
        return nums[m - 1][n - 1]
```
