```python
class Solution:
    def canJump(self, nums: List[int]) -> bool:
        
        max_reach = nums[0]
        
        for i in range(len(nums)):
            if i > max_reach:
                return False
            
            if i + nums[i] > max_reach:
                max_reach = i + nums[i]
            
        return True
        
```
