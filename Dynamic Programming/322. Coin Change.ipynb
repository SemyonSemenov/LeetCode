```python
def F(S, coins):
  if S == 0:
    return 0

  elif S > 0:
    return min([F(S - coin, coins) + 1 for coin in coins])
  
  else: return float('inf')
```

```python
class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:
        nums = [amount + 1] * (amount + 1)
        nums[0] = 0
        
        for i in range(1, len(nums)):
            for coin in coins:
                if i - coin >= 0:
                    nums[i] = min(nums[i - coin] + 1, nums[i])
        
        if nums[-1] == amount + 1:
            return -1
        else: return nums[-1]
```
