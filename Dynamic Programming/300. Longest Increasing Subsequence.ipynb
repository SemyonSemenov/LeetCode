```python
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        F = [1] * len(nums)

        for i in reversed(range(len(nums) - 1)):
            for j in range(i + 1, len(nums)):
                if nums[i] < nums[j]:
                    F[i] = max(F[i], F[j] + 1)
        return max(F)
```
