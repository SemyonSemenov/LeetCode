```python
from collections import deque

"""
# Definition for a Node.
class Node:
    def __init__(self, val: int = 0, left: 'Node' = None, right: 'Node' = None, next: 'Node' = None):
        self.val = val
        self.left = left
        self.right = right
        self.next = next
"""

class Solution:
    def connect(self, root: 'Node') -> 'Node':
        
        if root == None:
            return root 
        queue = deque([root])
        
        while queue:
            prev = None
            for _ in range(len(queue)):
                curr = queue.popleft()
                
                if curr.left != None:
                    queue.append(curr.left)
                if curr.right != None:
                    queue.append(curr.right)
                    
                if prev != None:
                    prev.next = curr
                
                prev = curr
                
                
        return root
```
