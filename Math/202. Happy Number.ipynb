```python
class Solution:
    def isHappy(self, n: int) -> bool:
        log = [n]

        while True:
            num = 0
            pos = 10

            for i in range(len(str(n))):
                digit = n % 10
                n = n // 10
                num += digit * digit

            if num in log:
                log.append(num)
                break
            log.append(num)

            n = num
        if log[-1] == 1:
            return True
        else:
            return False
```
