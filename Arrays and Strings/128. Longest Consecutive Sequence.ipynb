```python
class Solution(object):
    def longestConsecutive(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if nums == []:
            return 0
        a, b = min(nums), max(nums)
        nums = [x - a for x in nums]
        
        dim = b - a + 1
        C = [0] * (dim + 1)
        
        for obj in nums:
            C[obj] += 1
            
        answer = 0
        toWrite = True
        tmp = 0
        for obj in C:
            if obj > 0:
                toWrite = True
            if obj == 0:
                toWrite = False
                answer = max(tmp, answer)
                tmp = 0

            if toWrite == True and obj > 0:
                tmp += 1
            
        return answer
```
