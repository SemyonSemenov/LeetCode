```python
import numpy as np
from collections import Counter
```

```python
s = "ADOBECODEBANC"
t = "ABC"
```

```python
window = {}
dict_t = Counter(t)
n_unique = len(dict_t)
formed = 0 
answer = [float('inf'), 0, 0]
r, l = 0, 0


while r < len(s):

  character = s[r]
  window[character] = window.get(character, 0) + 1

  if character in dict_t and window[character] == dict_t[character]:
    formed += 1


  while l <= r and formed == n_unique:

    if r - l + 1 < answer[0]:
      answer = r - l + 1, l, r

    character = s[l]
    window[character] -= 1

    if character in dict_t and window[character] < dict_t[character]:
      formed -= 1

    l += 1
    
  r += 1
```

```python
class Solution(object):
    def minWindow(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        window = {}
        dict_t = Counter(t)
        n_unique = len(dict_t)
        formed = 0 
        answer = [float('inf'), 0, 0]
        r, l = 0, 0
        
        while r < len(s):

            character = s[r]
            window[character] = window.get(character, 0) + 1

            if character in dict_t and window[character] == dict_t[character]:
                formed += 1


            while l <= r and formed == n_unique:

                if r - l + 1 < answer[0]:
                    answer = r - l + 1, l, r

                character = s[l]
                window[character] -= 1

                if character in dict_t and window[character] < dict_t[character]:
                    formed -= 1

                l += 1
    
            r += 1
        return '' if answer[0] == float("inf") else s[answer[1] : answer[2] + 1]
```
