```python
buttons = [['a', 'b', 'c'], ['d', 'e', 'f'], ['g', 'h', 'i'], ['j', 'k', 'l'], ['m', 'n', 'o'], ['p',   'q', 'r', 's'], ['t', 'u', 'v'], ['w', 'x', 'y', 'z']]
digits = "23"

curr_buttons = []
for digit in digits:
  curr_buttons.append(buttons[int(digit) - 2])

helper = []
for i in range(len(digits)):
  helper.append([0, len((curr_buttons[i]))])
```

```python
ans = []
while True:

  tmp = ''
  for i in range(len(digits)):
    tmp += str(curr_buttons[i][helper[i][0]])
  ans.append(tmp)

  helper[-1][0] += 1
  if helper[-1][0] == helper[-1][1]:
      helper[-1][0] = 0
      helper[-2][0] += 1
  
  for j in reversed(range(1, len(helper) - 1)):
    if helper[j][0] == helper[j][1]:
      helper[j][0] = 0
      helper[j - 1][0] += 1
  
  if helper[0][0] == helper[0][1]:
    break
```

```python
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        buttons = [['a', 'b', 'c'], ['d', 'e', 'f'], ['g', 'h', 'i'], ['j', 'k', 'l'], ['m', 'n', 'o'], ['p',   'q', 'r', 's'], ['t', 'u', 'v'], ['w', 'x', 'y', 'z']]
            
        curr_buttons = []
        for digit in digits:
            curr_buttons.append(buttons[int(digit) - 2])

        helper = []
        for i in range(len(digits)):
            helper.append([0, len((curr_buttons[i]))])
        
        if len(digits) == 0:
            return []
        
        elif len(digits) == 1:
            return curr_buttons[0]
        ans = []
        while True:

            tmp = ''
            for i in range(len(digits)):
                tmp += str(curr_buttons[i][helper[i][0]])
            ans.append(tmp)

            helper[-1][0] += 1
            if helper[-1][0] == helper[-1][1]:
                helper[-1][0] = 0
                helper[-2][0] += 1
  
            for j in reversed(range(1, len(helper) - 1)):
                if helper[j][0] == helper[j][1]:
                    helper[j][0] = 0
                    helper[j - 1][0] += 1
  
            if helper[0][0] == helper[0][1]:
                break
            
        return ans
```
